### Before concatinating Two Dataset; Understanding the Structure of the DataFrames is Crucial, below key points to be considered to ensure a successful and accurate combination
### When preparing to concatenate two datasets, it is essential to consider the following factors to ensure a successful and accurate combination:
### Ensure Column Name Consistency:Verify that the column names in both DataFrames match. Discrepancies in naming can lead to confusion and errors in the resulting DataFrame.
### Check Data Type Compatibility: Confirm that the data types of corresponding columns are the same across both DataFrames. Mismatched data types can lead to unexpected results or errors during concatenation.
### Align Indices Properly: Ensure that the indices of both DataFrames are aligned. If the indices do not match, you may inadvertently combine data from different rows, resulting in inaccurate data representation and analysis.

### Important Note on ignore_index: concat_df = pd.concat([df2,df3],ignore_index = True )
### Without ignore_index= True : The resulting DataFrame may retain duplicate indices (e.g., indices 0 and 1 can appear multiple times).
### With ignore_index=True: keeps sequential index (0, 1, 2, 3), which is often more convenient for analysis and operations.



### Considering critical factors—consistent column names, compatible data types, and proper index alignment—ensures a smooth concatenation of df2 and df3. This attention to detail maintains data integrity and enables accurate analysis.


### The use of numeric values for weekdays, weathersit and unclear column names lacks clarity and fails to provide meaningful context.
### By adding a new columns with weekday_name, months_name,Weather_condition and renaming existing columns, we can enhance the user-friendly representation of the data.


```python
concat_df['Week_Name'] = concat_df['weekday'].replace(
    {0: 'Sun', 1: 'Mon', 2: 'Tue', 3: 'Wed', 4: 'Thu', 5: 'Fri', 6: 'Sat'}
)
```

```python
concat_df['Week_Name'].value_counts()
```

```python
concat_df['Month_Name'] = concat_df['mnth'].replace(
    {1: 'Jan', 2: 'Feb'})
```

## Note : The ref of Weather_Condition is taken from Overview pdf file

```python
concat_df['Weather_Condition'] = concat_df['weathersit'].replace(
    {1: 'Clear, Few clouds, Partly cloudy, Partly cloudy', 
     2: 'Mist + Cloudy, Mist + Broken clouds, Mist + Few clouds, Mist',
     3: 'Light Snow, Light Rain + Thunderstorm + Scattered clouds, Light Rain + Scattered clouds',
     4: 'Heavy Rain + Ice Pallets + Thunderstorm + Mist, Snow + Fog'
    })
```

```python
concat_df.info()
```

```python
concat_df.columns = [cap.capitalize() for cap in concat_df.columns]
```

```python
concat_df.columns
```

```python
concat_df.rename(columns={'Dteday': 'Datetime', 'Yr': 'Year', 'Mnth': 'Month', 'Hr': 'Hour','Weathersit':'Weather', 
                          'Temp':'Temperature','Hum':'Humidity', 'Cnt': 'Total_Count'}, inplace=True)
```


### The graphs below illustrate data distribution:
### The column [Weather] is represented in bar plot due to its four unique values. The chart highlights a significant distribution at value 1, which corresponds to weather conditions: Clear, Few clouds, Partly cloudy, and Overcast.


```python
fig, ax = plt.subplots(2,3, figsize=(20, 10)) 

sns.countplot(x=concat_df['Weather'],ax=ax[0,0],palette="coolwarm")
ax[0,0].set_title('Weather',fontsize=20, fontweight='bold',color="Blue")

sns.histplot(concat_df['Temperature'], kde=True,ax=ax[0, 1],color="royalblue") 
ax[0, 1].set_title('Temperature',fontsize=20, fontweight='bold',color="royalblue")

sns.histplot(concat_df['Atemp'], kde=True,ax=ax[0, 2],color="seagreen") 
ax[0, 2].set_title('Atemp',fontsize=20, fontweight='bold',color="seagreen")

sns.histplot(concat_df['Humidity'], kde=True,ax=ax[1,0],color="darkorange") 
ax[1,0].set_title('Humidity',fontsize=20, fontweight='bold',color="darkorange")

sns.histplot(concat_df['Windspeed'], kde=True,ax=ax[1,1],color="crimson") 
ax[1,1].set_title('Windspeed',fontsize=20, fontweight='bold',color="crimson")

sns.histplot(concat_df['Total_Count'],kde=True, ax=ax[1,2],color="purple") 
ax[1,2].set_title('Total_Count',fontsize=20, fontweight='bold',color="purple")

plt.tight_layout()
plt.show()
```

### The KDE indicates that the data for Temperature, Atemp, and Humidity follows a normal distribution, while Windspeed and Total Count exhibit right skewness.
